### GRAFICAR DEL 2014 LOS 5 NOMBRE QUE MAS SE REPITEN

```python
lista_años = dataset.anio.unique()
```

```python
for año in lista_años:
    print(dataset.loc[dataset.anio== 2014].sort_values('cantidad',ascending=False)\
          .head(), '\n')
```

```python
dataset_to_plot = dataset.groupby('nombre').sum().sort_values('cantidad',ascending=False).iloc[:5]
dataset_to_plot
```

```python

sns.barplot(x= dataset_to_plot.index,y='cantidad',data=dataset_to_plot );
```

```python
dataset_to_plotdes = dataset.groupby('nombre').sum().sort_values('cantidad',ascending=True).iloc[:5]
dataset_to_plotdes
```

```python
sns.barplot(x= dataset_to_plotdes.index,y='cantidad',data=dataset_to_plotdes );
```
